```python
#seriesReaderWriter01.py

#stpe01
import pandas as pd

myindex = ['서울', '부산', '광주', '대구', '울산', '목포', '여수']
mylist = [50, 60, 40, 80, 70, 30, 20]
myseries = pd.Series(data=mylist, index=myindex)
print(myseries)

#step02
print('\n색인의 이름으로 값 읽기')
print(myseries[['대구']])

print('\n라벨 이름이로 슬라이싱')
print(myseries['대구':'목포'])

#step03
print('\n여러 개의 색인 이름으로 데이터 읽기')
print(myseries[['대구', '여수']])

print('\n정수릉 이용한 데이터 읽기')
print(myseries[[2]])

#step04
print('\n0, 2, 4번째 데이터 읽기')
print(myseries[0:5:2])

#step05
myseries[2] = 22
myseries[2:5] = 33
myseries[['서울', '대구']] = 55
myseries[0::2] = 77

print(myseries)
```

```python
#2.3 데이터 프레임
import pandas as pd

sdata = {'city': ['서울', '서울', '서울', '부산', '부산'], \
         'year': [2000, 2001, 2002, 2001, 2002], \
         'pop': [1.5, 1.7, 3.6, 2.4, 2.9]}

mycolumn = ['city', 'year', 'pop']
myindex = ['one', 'two', 'three', 'four', 'five']
myframe = pd.DataFrame( sdata, columns=mycolumn, index=myindex)
print(myframe)
```

```python
import numpy as np
import pandas as pd

myindex = ['이순신', '김유신', '강감찬', '광해군', '연산군']
mycolumns = ['서울', '부산', '광주', '목포', '경주']
mylist = list( 10 * onedata for onedata in range(1, 26))
print(mylist)
print()

myframe = pd.DataFrame(np.reshape(mylist, (5,5)), index=myindex, columns=mycolumns)
print(myframe)
print()

print('\n#1 행만 Series로 읽어 오기')
result = myframe.iloc[1]
print(type(result))
print(result)
print()

print('\n#몇개의 행만 Series로 읽어 오기')
result = myframe.iloc[[1,3]]
print(type(result))
print(result)
print()

print('\n#짝수의 행만 Series로 읽어 오기')
result = myframe.iloc[0::2]
print(result)
print()

print('\n#이순 행만  Dataframe로 읽어 오기')
result = myframe.loc['이순신']
print(type(result))
print(result)
print()

print(myframe.index)
print('-'*40)

mytarget = np.random.choice(myframe.index, 3)
print(mytarget)

result = myframe.loc[['강감찬'], ['광주']]
print(result)

result = myframe.loc[myframe['부산'] <= 100]
print(result)
```
